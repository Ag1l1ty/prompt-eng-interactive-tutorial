### El SDK de Anthropic y la API de Messages
Utilizaremos el [SDK de Python de Anthropic](https://docs.anthropic.com/claude/reference/client-sdks) y la [API de Messages](https://docs.anthropic.com/claude/reference/messages_post) a lo largo de este tutorial.

A continuacion se muestra un ejemplo de como se vera la ejecucion de un prompt en este tutorial. Primero, creamos `get_completion`, que es una funcion auxiliar que envia un prompt a Claude y devuelve la respuesta generada por Claude. Ejecuta esa celda ahora.

```python
import anthropic

client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

Ahora escribiremos un prompt de ejemplo para Claude e imprimiremos la salida de Claude ejecutando nuestra funcion auxiliar `get_completion`. Al ejecutar la celda de abajo se mostrara una respuesta de Claude debajo de ella.

Sientete libre de modificar la cadena del prompt para obtener diferentes respuestas de Claude.

```python
# Prompt
prompt = "Hello, Claude!"

# Obtener la respuesta de Claude
print(get_completion(prompt))
```

Las variables `API_KEY` y `MODEL_NAME` definidas anteriormente se utilizaran a lo largo del tutorial. Solo asegurate de ejecutar las celdas de cada pagina del tutorial de arriba hacia abajo.
