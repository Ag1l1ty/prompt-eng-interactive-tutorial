# Apendice 10.3: Busqueda y Recuperacion

Sabias que puedes usar a Claude para **buscar en Wikipedia por ti**? Claude puede encontrar y recuperar articulos, momento en el cual tambien puedes usar a Claude para resumirlos y sintetizarlos, escribir contenido novedoso a partir de lo que encontro, y mucho mas. Y no solo Wikipedia! Tambien puedes buscar en tus propios documentos, ya sea almacenados como texto plano o incrustados en un almacen de datos vectoriales.

Consulta nuestros [ejemplos del cookbook de RAG](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb) para aprender como complementar el conocimiento de Claude y mejorar la precision y relevancia de las respuestas de Claude con datos recuperados de bases de datos vectoriales, Wikipedia, internet y mas. Alli tambien puedes aprender sobre como usar ciertas herramientas de [embeddings](https://docs.anthropic.com/claude/docs/embeddings) y bases de datos vectoriales.

Si te interesa aprender sobre arquitecturas avanzadas de RAG usando Claude, consulta nuestras [diapositivas de la presentacion tecnica de Claude 3 sobre arquitecturas RAG](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782).
